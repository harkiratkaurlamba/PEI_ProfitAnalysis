```python
def read_data_sink(header_flag, delimiter, inferSchema_flag, source_path, base_path , sub_folder_category, read_data_format):
    raw = spark.read.format(read_data_format).\
        option("inferSchema", inferSchema_flag).\
            option('header', header_flag).\
                load(source_path)
    sink_path = base_path + sub_folder_category
    print(sink_path)
    raw.write.mode("overwrite").parquet(sink_path)
    raw.show(2)

read_data_sink("true", ',', "false", '/Volumes/PEI/raw/data/Products.csv', '/Volumes/pei/bronze/','product/','csv')

#schema bhi bhejna h
```
